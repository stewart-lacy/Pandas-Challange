```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
len_of_data = purchase_data["SN"].nunique()
male=purchase_data.loc[purchase_data['Gender'] == "Male"]["SN"].nunique()
male_percent = ((male/len_of_data)*100)

female=purchase_data.loc[purchase_data['Gender'] == "Female"]["SN"].nunique()
female_percent = ((female/len_of_data)*100)

other= len_of_data - male - female
other_percent = ((other/len_of_data)*100)


Gender_Demographics_summary = pd.DataFrame({"": ["Male", "Female", "Other / Non-Disclosed"],"Total Count": [male,female,other], "Percentage of Players":[male_percent,female_percent, other_percent] })
Gender_Demographics_summary

Gender_Demographics_final= Gender_Demographics_summary.set_index("")
Gender_Demographics_final.style.format({"Percentage of Players": "{:.2f}%"})
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
len_of_purchase = purchase_data["Purchase ID"].nunique()
avg_purchase_price = purchase_data["Price"].nunique()
#Purchase Count

female_purchase_count=purchase_data.loc[purchase_data['Gender'] == "Female"]["Purchase ID"].nunique()
male_purchase_count=purchase_data.loc[purchase_data['Gender'] == "Male"]["Purchase ID"].nunique()
other_purchase_count = (len_of_purchase - male_purchase_count - female_purchase_count)
#Average Puchase Price

female_avg_purchase_price = purchase_data.loc[purchase_data['Gender'] == "Female"]["Price"].mean()
male_avg_purchase_price = purchase_data.loc[purchase_data['Gender'] == "Male"]["Price"].mean()
other_avg_purchase_price = purchase_data.loc[purchase_data['Gender'] == "Other / Non-Disclosed"]["Price"].mean()

#Total Purchase Value 
female_sum_purchase_price = purchase_data.loc[purchase_data['Gender'] == "Female"]["Price"].sum()
male_sum_purchase_price = purchase_data.loc[purchase_data['Gender'] == "Male"]["Price"].sum()
other_sum_purchase_price = purchase_data.loc[purchase_data['Gender'] == "Other / Non-Disclosed"]["Price"].sum()

#Average Total Purchase per Person
 #avg_female_purchase_count= 
avg_total_female = (female_sum_purchase_price/female)
avg_total_male = (male_sum_purchase_price/male)
avg_other = (other_sum_purchase_price/other)


pd.DataFrame({"Gender": ["Female", "Male", "Other / Non-Disclosed"],"Purchase Count": [female_purchase_count,male_purchase_count,other_purchase_count], 
              "Average Purchase Price":[female_avg_purchase_price,male_avg_purchase_price, other_avg_purchase_price],
              "Total Purchase Value":[female_sum_purchase_price,male_sum_purchase_price,other_sum_purchase_price], 
              "Avg Total Purchase per Person": [avg_total_female,avg_total_male,avg_other]
             })
```
