## Gender Demographics

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
load_file_1 = "Resources/purchase_data.csv"
new_purchase_data = pd.read_csv(load_file_1)
new_purchase_data
```

```python
bins = [-1,9.99,14.99,19.99,24.99,29.99,34.99,39.99,1000]
category =["<10","10-14","15-19","20-24","25-29","30-34","35-39","40 +" ]

#Categorize the existing players using the age bins. Hint: use pd.cut()
test_3 = new_purchase_data
test_3["SN"] = pd.cut(test_3["Age"],bins, labels=category)
SN_grouped = test_3.groupby(["SN"])

#Purchase count 
bin_count = age_grouped["Age Grouped"].count()
#avg. purchase price
average_price = age_grouped["Price"].mean()
#Total purchase value
total_price = age_grouped["Price"].sum()
#Average total
new_data = (age_grouped["Price"].sum()/SN_grouped["SN"].count())

purchase_analysis = pd.DataFrame({"Purchase Count": bin_count,
                               "Average Purchase Price": average_price,
                                "Total Purchase Value":total_price, 
                                  "Avg Total Purchase per Person": new_data})
purchase_analysis["Average Purchase Price"] = purchase_analysis["Average Purchase Price"].map("$ {:,.2f}".format)
purchase_analysis["Total Purchase Value"] = purchase_analysis["Total Purchase Value"].map("$ {:,.2f}".format)
purchase_analysis["Avg Total Purchase per Person"] = purchase_analysis["Avg Total Purchase per Person"].map("$ {:,.2f}".format)
purchase_analysis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
load_file_2 = "Resources/purchase_data.csv"
new_purchase_data_1 = pd.read_csv(load_file_2)
new_purchase_data_1
```

```python
grouped_SN = new_purchase_data_1.groupby("SN")
count_SNgrouped = grouped_SN["Item ID"].count()
count_SNTotal = grouped_SN["Price"].sum()
average_price = (count_SNTotal/count_SNgrouped)
top_spenders = pd.DataFrame({"Purchase Count": count_SNgrouped, "Average Purchase Price":average_price, "Total Purchase Value":count_SNTotal} )
top_spenders = top_spenders.sort_values("Total Purchase Value", ascending = False)
top_spenders["Average Purchase Price"] = top_spenders["Average Purchase Price"].map("${:.2f}".format)
top_spenders["Total Purchase Value"] = top_spenders["Total Purchase Value"].map("${:.2f}".format)

top_spenders.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_item=new_purchase_data_1.groupby(["Item ID", "Item Name"])
count_item_id = popular_item["SN"].count()
count_purchase_price = popular_item["Price"].sum()
sum_total_purchase = (count_purchase_price/count_item_id)
price_total = (sum_total_purchase * count_item_id)
Top_five_item = pd.DataFrame({"Purchase Count": count_item_id, "Item Price":sum_total_purchase, "Total Purchase Value": price_total})
Top_five_item = Top_five_item.sort_values ("Purchase Count", ascending = False)
Top_five_item["Item Price"] = Top_five_item["Item Price"].map("${:.2f}".format)
Top_five_item["Total Purchase Value"] = Top_five_item["Total Purchase Value"].map("${:.2f}".format)

Top_five_item.head(5)
```

## Most Profitable Items

```python
popular_item=new_purchase_data_1.groupby(["Item ID", "Item Name"])
count_item_id = popular_item["SN"].count()
count_purchase_price = popular_item["Price"].sum()
sum_total_purchase = (count_purchase_price/count_item_id)
price_total = (sum_total_purchase * count_item_id)
Top_five_item = pd.DataFrame({"Purchase Count": count_item_id, "Item Price":sum_total_purchase, "Total Purchase Value": price_total})
Top_five_item = Top_five_item.sort_values ("Total Purchase Value", ascending = False)
Top_five_item["Item Price"] = Top_five_item["Item Price"].map("${:.2f}".format)
Top_five_item["Total Purchase Value"] = Top_five_item["Total Purchase Value"].map("${:.2f}".format)

Top_five_item.head(5)
```
